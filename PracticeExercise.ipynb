```python
import pyspark
from pyspark.sql import SparkSession
```

```python
from pyspark.core.context import SparkContext
spark = SparkSession.builder.appName("ParallelizeExample").getOrCreate()
sparkContext  = spark.sparkContext

rdd = sparkContext.parallelize ([1,2,3,4,5,6,7])
rddCollect = rdd.collect()
print("Number  of Partitions: "+ str(rdd.getNumPartitions()))
print("Action: First element: "+ str(rdd.first()))
print(rddCollect)

```

```python
# Force Spark to use 4 partitions instead of 2
rdd_four = sparkContext.parallelize([1,2,3,4,5,6,7], 6)

print("New Number of Partitions: " + str(rdd_four.getNumPartitions()))

# Use glom() to actually see how the data is split across those 4 partitions
print("Partition distribution: " + str(rdd_four.glom().collect()))
```

```python
data = [2,6,7,6,9,4,2,0]
data_rdd = spark.sparkContext.parallelize(data)
data_rdd.reduce(lambda a, b: a * b)
```
